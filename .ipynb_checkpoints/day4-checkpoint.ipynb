```python
import numpy as np
import pandas as  pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
from scipy.stats import spearmanr
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data,columns = data.feature_names)
df['target'] = data.target
```

```python
corr = df.corr(method='pearson').abs()
```

```python
sorter = corr['target'].sort_values(ascending=False)
sorter.head(11)
```

```python
correlated = df.drop(['worst radius','concave points error','perimeter error','area error','radius error','mean radius','mean radius','mean texture','mean perimeter','mean area','mean compactness','mean concavity','mean concave points','worst texture','worst perimeter','worst area','worst smoothness','worst compactness','worst concavity','worst concave points','worst symmetry'],axis=1)
```

```python
correlated
```

```python
top_5 = correlated.drop(['mean smoothness','mean symmetry','compactness error','concavity error','worst fractal dimension','target'],axis=1)
top_5
```

```python
sns.pairplot(top_5)
```
